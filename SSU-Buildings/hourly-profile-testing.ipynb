This notebook shows an approach to generating average profiles from a time series.
The groupby command tells the computer to extract each sample with the same hour and minute and place them in a "group".
Once you have that group, you can perform statistics on it.

Right now, I'm showing how to get the mean, median, max, and min.
It would be better to plot each hour or min as a box plot but I don't have that yet.

```python
import pandas as pd
schulz = pd.read_csv('Schulz_clean_Nov.csv', parse_dates=True, index_col=0)
```

```python
%matplotlib inline
mean = schulz.groupby([schulz.index.hour, schulz.index.minute]).mean()
median = schulz.groupby([schulz.index.hour, schulz.index.minute]).median()
maximum = schulz.groupby([schulz.index.hour, schulz.index.minute]).max()
minimum = schulz.groupby([schulz.index.hour, schulz.index.minute]).min()
```

```python
mean.plot()
median.plot()
maximum.plot()
minimum.plot()
```
