# Degree Day Example

```python
%matplotlib inline
import pandas as pd

etc = pd.read_csv('ETC-Weather-2016.csv', index_col=0, parse_dates=True)
```

```python
# plot how much warmer the building is than outside

etc['deltaT'] = etc['Inside Temp (C)'] - etc['Outside Temp (C)']
etc['deltaT'].plot()
```

```python
# uses the cumulative sum function to add up all the delta Ts and get degree hours
# divide by 24 to get degree days
# only looking at December

(etc['deltaT']['2016-12-01':'2017-01-01'].cumsum()/24).plot()
```

```python
# you can also look at the cooling degree days by only adding the negative

(etc['deltaT'][etc['deltaT']<0].cumsum()/24).plot()
```
